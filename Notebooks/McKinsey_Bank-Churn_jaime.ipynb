```python
num_cont_end = data['contract_end'].notna().sum()
num_rows = data.shape[0]
print("Number of clients who HAVE ENDED their contracts: " + str(num_cont_end))
print("Percent of the total: " + str(100*num_cont_end/num_rows) + "%")
```

```python
# Add new binary column 'churn', this is the LABEL column
# 0 -> Client HAS NOT churn
# 1 -> Client HAS churn
churn=data['contract_end'].notna().astype(int)
data['churn']=churn
data
```

```python
# Add two new columns 'age' and 'contract_duration' 
# 'contract_duration' depends on whether the customer has finished his contract or not,
# so we select the min value between (today - start) and (today - end)
today = pd.to_datetime('2019-10-16')
data['age'] = today-pd.to_datetime(data['date_of_birth'])
data['contract_start_to_date_time'] = today-pd.to_datetime(data['contract_start'])
data['contract_end_to_date_time'] = today-pd.to_datetime(data['contract_end'])
data['contract_duration'] = data.loc[:, ['contract_start_to_date_time', 'contract_end_to_date_time']].min(axis=1)
# Drop un used columns
data=data.drop(['contract_start','contract_end','date_of_birth','ZIP','profession','consumer_credit_maturity',
                'consumer_credit_value','cash_withdrawals_value','customer_limit','mortgage_numbers',
                'mortgage_dayuntilmaturity','mortgage_downpayment','mortgage_interest','mortgage_value',
                'insurance_car_premium_per_year','insurance_house_premium_per_year','income_securities_per_year',
                'insurance_life_premium_per_month','contract_end_to_date_time','income_salary_per_year',
                'last_balance_minus_6_months','last_balance_minus_12_months','no_credit_rating_flag',
                'income_deposits_per_year','contract_start_to_date_time'],axis=1)
data
```
